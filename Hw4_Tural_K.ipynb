3) Get the exchange rate HUF vs USD from 2000 to 2010, but every six months. Plot it on a plot.

```python
startdt=datetime.date(2000,1,1)
enddt=datetime.date(2011,12,1)
intrvl = enddt-startdt
dlt=intrvl.days/180


dlt = math.floor(dlt)
dlt=int(dlt)

time2data6={}
m6keys=[]
l6values=[]
for i in range(0, dlt+1):
    nepp6=startdt+relativedelta(months=i*6)
    URL = "%s/historical/2011-10-18.json?%s" % (base_url, id_str)
    request = urllib.request.Request(URL)
    result = urllib.request.urlopen(request)
    text = result.read()
    data = json.loads(text)
    t = datetime.datetime.fromtimestamp(data["timestamp"])
    unv=data
    unk=(nepp6.year,nepp6.month)
    m6keys.append(unk)
    l6values.append(unv)
    time2data6={key:value for key, value in zip(m6keys,l6values)}
time2data6
```

```python
list_ex = []
list_ts = []

for y_m in sorted(time2data6.keys()): # sorted() makes sure we loop
                                     # over the dict in time order
    list_ts.append(y_m)
    
    data = time2data6[y_m]
    exch_eur = data["rates"]["HUF"]
    list_ex.append(exch_eur)
```

```python
list_dt = []
for yr,mo in list_ts:
    dt = datetime.datetime(year=yr,month=mo,day=1)
    list_dt.append(dt)
```
